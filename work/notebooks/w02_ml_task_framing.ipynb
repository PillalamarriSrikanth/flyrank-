## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

Simple static rules (like if impressions drop > 20%) are too fragile. They fail to separate normal seasonal traffic variations, global site-wide tracking updates, or minor volatility in Google's indexing positions. A machine learning model automatically maps non-linear combinations of clicks, baseline traffic volumes, and rank changes to evaluate context far more effectively than rigid, manual heuristics.
